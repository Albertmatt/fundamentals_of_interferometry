## 3.5 Horizontal Coordinates (ALT,AZ)

In Sec. [EXREF](../3_Positional_Astronomy/C_Hour_Angle.ipynb#pos:sec:ha) <!--\ref{pos:sec:ha}--> we introduced the concept of an hour angle, which allows us to determine the time that still needs to elapse before a source crosses the local meridian. This however does not tell us where we should point a telescope on earth inorder to observe a source with a specific hour angle. The horizontal coordinates azimuth $A$ and altitude $E$ (elevation) is used to enable an observer on earth to locate celestial objects in the observers local sky.
The observer's horizontal plane is the fundamental plane of this coordinate system and is known as the celestial horizon. The azimuth angle is measured in the fundamental plane from due north towards the east, while the altitude of a celestial object is the angle between it and the observer's local horizon. Both azimuth and elevation is measured in degrees. The azimuth and elevation angle is depicted in Fig. [INREF](#pos:fig:horizontal) <!--\ref{pos:fig:horizontal}-->

The equations below allow us to convert between equatorial and horizontal coordinates

\begin{eqnarray}
\cos\delta\cos H &=& \cos L\sin E - \sin L\cos E\cos A\\
-\cos\delta\sin H&=& \cos E\sin A\\
\sin\delta &=& \sin L\sin E+\cos L \cos E \cos A 
\end{eqnarray}

These equations were derived by applying the spherical trigonometric identities in Sec. [EXREF](../3_Positional_Astronomy/A_Spherical_Trigonometry.ipynb#pos:sec:st}) <!--\ref{pos:sec:st}--> to 
the triangle $\Delta PSZ$ which is depicted in Fig. [INREF](#pos:fig:conversion_alaz_radec) <!--\ref{pos:fig:conversion_alaz_radec}-->.
