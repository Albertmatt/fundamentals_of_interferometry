## 3.2 Equatorial Coordinates (RA,DEC)

We can use a geographical coordinate system to uniquely identify a position on earth. We normally use the coordinates *latitude* $L$ (to measure north and south) and *longitude* $\mathcal{L}$ (to measure east and west) to accomplish this. 

We also require a coordinate system to locate the stars. For practical purposes we may think of our universe as being projected onto a sphere of arbitrary radius. This sphere surrounds the Earth and is known as the *celestial sphere*. This is not a true representation of our universe, but it is a very usefull construct that we use to keep track of the stars in the night sky. The *north celestial pole* (NCP) is an important location on the celestial sphere and is obtained by projecting the north pole of the earth onto the celestial sphere. The star *Polaris* is very close to the NCP. The *south celestial pole* (SCP) is obtained in a similar way. The imaginary circle known as the *celestial equator* is in the same plane as the equator of the earth and is obtained by projecting the equator of the earth onto the celestial sphere.
We use a specific point on the celestial equator from which we measure the location of all other celestial objects. This point is know as the *first point of Airies* <!--$\vernal--> or the *vernal equinox*. The physical significance of this point is discussed in Sec. [EXREF](../3_Positional_Astronomy/D_Local_Sidereal_Time.ipynb#pos:sec:lst) <!--\ref{pos:sec:lst}-->.  

The stars themselves do not move on the celestial sphere and therefore have a unique location on it (at least not much in one generation). We use the *equatorial coordinates*, *right ascension* $\alpha$ (measured in hours) and *declination* $\delta$ (measured in degrees) to uniquely identify the location of celestial objects. The *hour circle* of an object is the circle on the celestial sphere that crosses the NCP and the object itself and is perpendicular to the celestial equator. The right ascension of an object is the angular distance between the vernal equinox and the hour circle of a celestial object measured along the celestial equator and is measured in an easterly direction. The declination of an object is the  angular distance it is away from the celestial equator measured along its hour circle (it is positive in the northern celestial hemisphere and negative in the southern celestial hemisphere). ### 3.2.1 The Terestial Reference Frame (X,Y,Z)

We use the $XYZ$-terestial reference frame as an intermediate coordinate system when 
we calculate our instrument's $uv$-coverage. We can denote the orientation of the $XYZ$-coordinate system by using equatorial coordinates $(H; \delta)$, with $H$ being
the hour angle and $\delta$ the declination. In the coordinate system $XYZ$, the $X$-axis points towards $(0\textrm{h}; 0^{\circ})$ (the point where the vernal equinox crosses the local meridian), the $Y$-axis towards $(-6\textrm{h}; 0^{\circ})$ (due east) and the $Z$-axis towards the North Celestial Pole (NCP). Note that althoug we use equatorial 
coordinates to define the $XYZ$-coordinate system it represents a terestial reference frame, as the $X$, $Y$ and $Z$ axes never change orientation for a local observer on earth. The antenna positions of an array are normally given in ENU (east-north-up) coordinates. The first thing we do is to use the ENU coordinates of our array to calculate baseline difference vectors. We then obtain the magnitude, the azimuth angle and the elevation angle of these difference vectors. We then use the magnitude $s$, the azimuth angle $A$ and the elevation angle $E$ of the difference vectors as well as the latitude $L$ of the array to convert our baseline vectors into the XYZ-coordinate system. The following equation describes this conversion in a mathematical rigorous way.     

\begin{equation}
\begin{bmatrix}
X\\
Y\\
Z
\end{bmatrix}
=
\begin{bmatrix}
s\cos \delta \cos H\\
-s\cos \delta \sin H\\
s\sin \delta
\end{bmatrix}
=s
\begin{bmatrix}
\cos L \sin E - \sin L \cos E \cos A\nonumber\\ 
\cos E \sin A \nonumber\\
\sin L \sin E + \cos L \cos E \cos A
\end{bmatrix}
\end{equation}

The above equation can be obtained by applying the spherical trigonometric identities in Sec. [EXREF](../3_Positional_Astronomy/A_Spherical_Trigonometry.ipynb#pos:sec:st}) <!--\ref{pos:sec:st}--> to 
the triangle $\Delta ZTS$ which is depicted in Fig. [INREF](#vis:fig:conversion_xyz_XYZ) <!--\ref{vis:fig:conversion_xyz_XYZ}-->.
